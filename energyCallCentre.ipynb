```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
"""
Reading the file EnergyCallCentre.csv and converting it to a dataframe
"""

file = 'EnergyCallCentre.csv'
df = pd.read_csv(file)
```

```python
"""
Exploring the dataframe (EnergyCallCentre) for :
1. Shape
2. Columns
3. Index Values
4. First 5 rows
"""

filename = file.strip('.csv')
print("Shape of the dataframe (", filename, ") is: ", df.shape ,"\n")
print("Column names in the dataframe (", filename, ") is: \n", df.columns, "\n")
print("Index values of the dataframe (", filename, ") is: ", df.index, "\n")
print(df.head())
```

```python
"""
1. Exploring the dataframe (EnergyCallCentre) for : Null Values
2. Changing 'afternoo' to 'afternoon' in ToD column of the dataframe
"""

print(df.isna().any())

for lab, col in df.iterrows():
    if df.loc[lab,'ToD'] == 'afternoo':
        df.loc[lab,'ToD'] = 'afternoon'
df.head()
```

```python
fig, ax = plt.subplots()
numAgentsByTod = df.groupby('ToD').agg({'Agents':'sum'})
sortedNumAgentsByTod = numAgentsByTod.sort_values('Agents', ascending = False)
print(sortedNumAgentsByTod, "\n")
ax.bar(sortedNumAgentsByTod.index,sortedNumAgentsByTod['Agents'], color = 'orange')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Number of Agents')
ax.set_title('Agents active during the time of day')

fig, ax = plt.subplots()
numAgentsByMonth = df.groupby('Month').agg({'Agents':'sum'})
sortedNumAgentsByMonth = numAgentsByMonth.sort_values('Agents', ascending = False)
print(sortedNumAgentsByMonth)
ax.bar(sortedNumAgentsByMonth.index,sortedNumAgentsByMonth['Agents'], color = 'orange')
ax.set_xlabel('Month')
ax.set_ylabel('Number of Agents')
ax.set_title('Agents active during the three 2-month long periods of time')
```

```python
fig, ax = plt.subplots()
numCallsOfferedByTod = df.groupby('ToD').agg({'CallsOffered':'sum'})
sortedNumCallsOfferedByTod = numCallsOfferedByTod.sort_values('CallsOffered', ascending = False)
print(sortedNumCallsOfferedByTod, "\n")
ax.bar(sortedNumCallsOfferedByTod.index, sortedNumCallsOfferedByTod['CallsOffered'])
ax.set_xlabel('Time of Day')
ax.set_ylabel('Calls Offered')
ax.set_title('Calls Offered during the time of day')

fig, ax = plt.subplots()
numCallsOfferedByMonth = df.groupby('Month').agg({'CallsOffered':'sum'})
sortedNumCallsOfferedByMonth = numCallsOfferedByMonth.sort_values('CallsOffered', ascending = False)
print(sortedNumCallsOfferedByMonth)
ax.bar(sortedNumCallsOfferedByMonth.index, sortedNumCallsOfferedByMonth['CallsOffered'])
ax.set_xlabel('Month')
ax.set_ylabel('Calls Offered')
ax.set_title('Calls Offered during the three 2-month long periods of time')
```

```python
fig, ax = plt.subplots()
numCallsHandledByToD = df.groupby('ToD').agg({'CallsHandled':'sum'})
sortedNumCallsHandledByToD = numCallsHandledByToD.sort_values('CallsHandled', ascending = False)
print(sortedNumCallsHandledByToD, "\n")
ax.bar(sortedNumCallsHandledByToD.index, sortedNumCallsHandledByToD['CallsHandled'], color = 'green')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Calls Handled')
ax.set_title('Calls Handled during the time of day')

fig, ax = plt.subplots()
numCallsHandledByMonth = df.groupby('Month').agg({'CallsHandled':'sum'})
sortedNumCallsHandledByMonth = numCallsHandledByMonth.sort_values('CallsHandled', ascending = False)
print(sortedNumCallsHandledByMonth)
ax.bar(sortedNumCallsHandledByMonth.index, sortedNumCallsHandledByMonth['CallsHandled'], color = 'green')
ax.set_xlabel('Month')
ax.set_ylabel('Calls Handled')
ax.set_title('Calls Handled during the three 2-month long periods of time')
```

```python
fig, ax = plt.subplots()
numCallsAbandonedByToD = df.groupby('ToD').agg({'CallsAbandoned':'sum'})
sortedNumCallsAbandonedByToD = numCallsAbandonedByToD.sort_values('CallsAbandoned', ascending = False)
print(sortedNumCallsAbandonedByToD, "\n")
ax.bar(sortedNumCallsAbandonedByToD.index, sortedNumCallsAbandonedByToD['CallsAbandoned'], color = 'red')
ax.set_xlabel('Time of Day')
ax.set_ylabel('Calls Abandoned')
ax.set_title('Calls Abandoned during the time of day')

fig, ax = plt.subplots()
numCallsAbandonedByMonth = df.groupby('Month').agg({'CallsAbandoned':'sum'})
sortedNumCallsAbandonedByMonth = numCallsAbandonedByMonth.sort_values('CallsAbandoned', ascending = False)
print(sortedNumCallsAbandonedByMonth)
ax.bar(sortedNumCallsAbandonedByMonth.index, sortedNumCallsAbandonedByMonth['CallsAbandoned'], color = 'red')
ax.set_xlabel('Month')
ax.set_ylabel('Calls Abandoned')
ax.set_title('Calls Abandoned during the three 2-month long periods of time')
```
